### Importing packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Loading the data

```python
data=np.loadtxt('Audiobooks_data.csv',delimiter=',')
```

```python
inputs=data[:,:-1]
target=data[:,-1:]
```

### balancing the data

```python
target_sum=int(np.sum(target))
zero_sum=0
indices_to_remove=[]
for i in range(target.shape[0]):
    if target[i]==0:
        zero_sum +=1
        if zero_sum>target_sum:
            indices_to_remove.append(i)
```

```python
balanced_target=np.delete(target,indices_to_remove,axis=0)
balanced_input=np.delete(inputs,indices_to_remove,axis=0)
```

### Scaling the data

```python
from sklearn import preprocessing
```

```python
scaled_input=preprocessing.scale(balanced_input)
```

### Shuffling the data

```python
shuffle_data=np.arange(scaled_input.shape[0])
np.random.shuffle(shuffle_data)

shuffle_target=balanced_target[shuffle_data]
shuffle_input=balanced_input[shuffle_data]
```

### Batching the data

```python
def batch_data(inputs,targets,per1,per2):
    import numpy as np
    global train_input
    global train_target
    global valid_input
    global valid_target
    global test_input
    global test_target
    count=inputs.shape[0]
    train_count=int(per1*count)
    valid_count=int(per2*count)
    test_count=count-train_count-valid_count
    train_input=inputs[:train_count]
    train_target=targets[:train_count]
    valid_input=inputs[train_count:train_count+valid_count]
    valid_target=targets[train_count:train_count+valid_count]
    test_input=inputs[train_count+valid_count:]
    test_target=targets[train_count+valid_count:]
    return train_input,train_target,valid_input,valid_target,test_input,test_target
   
   
    
```

```python
batch_data(shuffle_input,shuffle_target,0.75,0.15)
```
